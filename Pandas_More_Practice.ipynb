# 1- Write a Pandas program to build DataFrame from the following lists

```python
names = ['Ahmad', 'Fatma', 'Mahmoud', 'Anas', 'Omar', 'Sayed', 'Ali', 'Maryam', 'Habiba', 'Alaa']
scores = [12.5, 9, 16.5, 17, 9, 20, 14.5, 10, 8, 19]
attempts = [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
qualify = ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
```

```python
import pandas as pd


exam_data  = {'name': names,
        'score': scores,
        'attempts': attempts,
        'qualify': qualify}
df = pd.DataFrame(exam_data)
df   
```

# 8- Write a pandas program to find the minimum, maximum, range, Q1, Q2, Q3 and IQR of scores. 

```python
min_score = df.score.min()
max_score = df.score.max()
q1 = df.score.quantile(0.25)
q2 = df.score.quantile(0.50)
q3 = df.score.quantile(0.75)

score_range = max_score - min_score
iqr = q3 - q1

print('Min. Score: ', min_score)
print('Max. Score: ', max_score)
print('Score Range: ', score_range)
print('First Quantile: ', q1)
print('Second Quantile (Median): ', q2)
print('Third Quantile: ', q3)
print('Score Interquartile: ', iqr)
```
